```python
from qiskit import QuantumCircuit
import matplotlib.pyplot as plt

def My_Tof():
    qc = QuantumCircuit(3)

    qc.t(0)
    qc.h(2)

    qc.cx(0,1)
    qc.tdg(1)
    qc.cx(0,1)
    qc.t(1)

    qc.cx(1,2)
    qc.tdg(2)
    qc.cx(0,2)
    qc.t(2)
    qc.cx(1,2)
    qc.tdg(2)
    qc.cx(0,2)
    qc.t(2)
    qc.h(2)

    return qc


qc1 = My_Tof()
qc1.draw('mpl')
```

```python
def qiskit_Tof():
    qc = QuantumCircuit(3, name = "QiskitTof")
    qc.ccx(0,1,2)
    return qc

qc2 = qiskit_Tof()
qc2.decompose().draw("mpl")
```

```python
from qiskit.quantum_info import Operator
print(Operator(qc).equiv(Operator(qc2)))

print("My Toffoli:", qc1.count_ops())
print("Qiskit CCX:", qc2.decompose().count_ops())
print("My Toffoli depth:", qc1.depth())
print("Qiskit CCX depth:", qc2.decompose().depth())


```
